```python
import os
import pickle
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np 

with open('trainHistoryDict', "rb") as file_pi:
    history = pickle.load(file_pi)

from keras.models import load_model
# To load the model later
loaded_model = load_model('model.h5')
```

```python
# Extracting accuracy and loss data from the training history
accuracy = history['accuracy']  # Training accuracy per epoch
val_accuracy = history['val_accuracy']  # Validation accuracy per epoch

loss = history['loss']  # Training loss per epoch
val_loss = history['val_loss']  # Validation loss per epoch

# Generating a range object for the number of epochs
epochs = range(len(accuracy))

# Plotting Training Accuracy and Loss
plt.figure()  # Creating a new figure
plt.plot(epochs, accuracy, label='Training Accuracy')  # Plotting training accuracy
plt.plot(epochs, loss, label='Training Loss')  # Plotting training loss
plt.legend()  # Adding a legend to the plot
plt.title('Training Accuracy and Loss')  # Setting the title of the plot

# Plotting Validation Accuracy and Loss
plt.figure()  # Creating another figure for validation data
plt.plot(epochs, val_accuracy, label='Validation Accuracy')  # Plotting validation accuracy
plt.plot(epochs, val_loss, label='Validation Loss')  # Plotting validation loss
plt.legend()  # Adding a legend to this plot as well
plt.title('Validation Accuracy and Loss')  # Setting the title of the plot

plt.show()  # Displaying the plots
```
